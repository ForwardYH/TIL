### 교재와 동영상 강의를 참고하여 공부한 파일입니다.

* 강의 영상
    * https://www.youtube.com/playlist?list=PLBXuLgInP-5m_vn9ycXHRl7hlsd1huqmS

```python
'''https://search.naver.com/search.naver
   ?where=nexearch
   &sm=top_hty
   &fbm=1
   &ie=utf8
   &query=%EC%B4%88%EC%BD%9C%EB%A6%BF

웹 브라우저에 주소를 입력했기 때문에 GET방식
방식 : Get
대상 : search.naver.com => 호스트 이름
추가적인 정보 :
- 경로 : /search.naver
- 데이터 :
        ?where=nexearch
        &sm=top_hty
        &fbm=1
        &ie=utf8
        &query=초콜릿 # %EC%B4%88%EC%BD%9C%EB%A6%BF (인코딩 된 문자)
                        <= 디코딩 된 것이 초콜릿'''
```

```python
import urllib.request
import urllib.parse

api = "https://search.naver.com/search.naver"
values ={
    "where": "nexearch",
    "sm": "toy_hty",
    "fbm": "1",
    "ie": "utf8",
    "query": "초콜릿"
}
params = urllib.parse.urlencode(values)
#urllib의 parse를 사용한 이유는 초콜릿을 뜻하는 문자로 인코딩하기 위해 사용
url = api + "?" + params

data = urllib.request.urlopen(url).read()
text = data.decode("utf-8") # 한글을 표기하기 위해 utf-8로 디코딩. 
                            # euc-kr 테스트 해보기
print(text)
```
