```python
import requests
from bs4 import BeautifulSoup
from collections import Counter
```

# 호출가능한 객체

```python
class Calculator2:
    def __init__(self, base):
        self.base = base

    def __call__(self, x, y):
        return self.base + x + y
```

```python
calc2 = Calculator2(10)
```

```python
calc2.__call__(10, 20)
```

- 인스턴스 함수 하려면  언더스코어(__ call __) 멤버함수 구현

```python
calc2(10, 20)
```

```python
import re
```

```python
def word_count(url):
    html = requests.get(url).text
    soup = BeautifulSoup(html, 'html.parser')
    words = soup.text.split()
    
    words = [word for word in words if re.match(r'^[ㄱ-힣]+$', word)] # 한글만 가져오기 정규표현식 사용
    counter = Counter(words)
    
    return counter
word_count('https://nomade.kr/vod/')
```

```python
class WordCount:
    def get_text(self, url):
        '문자열 수집'
        html = requests.get(url).text
        soup = BeautifulSoup(html, 'html.parser')
        return soup.text
    
    def get_list(self, text):
        ' 단어 리스트로 변환'
        return text.split()
    
    def __call__(self, url):     # 호출 가능하도록 __call__ 함수
        text = self.get_text(url)
        words = self.get_list(text)
        counter = Counter(words)
        return counter

class KoreanWordCount(WordCount):# 상속과 오버라이딩을 통해 주요코드만 재정의
    def get_list(self, text):    # 이 함수는 부모 클래스의 get_list 함수를 오버라이딩 한것.
        words = text.split()
        korean_words = [word for word in words if re.match(r'^[ㄱ-힣]+$', word)]
        return korean_words
```

```python
korean_word_count = KoreanWordCount()
korean_word_count('https://nomade.kr/vod/')
```
